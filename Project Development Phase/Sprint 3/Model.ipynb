Image Augmentation / PreProcessing :

```python
#Import req. Lib.
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#Augmentation On Training Variable
train_datagen = ImageDataGenerator(rescale= 1./255,
                 zoom_range=0.2,
                 horizontal_flip =True)
```

```python
#Augmentation On Testing Variable
test_datagen = ImageDataGenerator(rescale= 1./255)
```

```python
#Augmentation On Training Variable
ftrain = train_datagen.flow_from_directory('/content/Dataset_Collection/Train',
                                           target_size=(64,64),
                                           class_mode='categorical',
                                           batch_size=100)
```

```python
#Augmentation On Training Variable
ftest = test_datagen.flow_from_directory('/content/Dataset_Collection/Test',
                                          target_size=(64,64),
                                          class_mode='categorical',
                                          batch_size=100)
```

**Model Building**

Adding Layers:

```python
#Import req. Lib.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```python
# Build a CNN Block:
model = Sequential() #intializing sequential model
model.add(Convolution2D(32,(3,3),activation='relu', input_shape=(64,64,3))) #convolution layer
model.add(MaxPooling2D(pool_size=(2, 2))) #Maxpooling layer
model.add(Flatten()) #Flatten layer
model.add(Dense(400,activation='relu')) #Hidden Layer 1
model.add(Dense(200,activation='relu')) #Hidden Layer 2
model.add(Dense(5,activation='softmax')) #Output Layer
```

Compiling

```python
# Compiling The Model...
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

Fit/Train The Model

```python
#Train Model:
model.fit_generator(ftrain,
                    steps_per_epoch=len(ftrain),
                    epochs=10,
                    validation_data=ftest,
                    validation_steps=len(ftest))
```

Saving The Model :

```python
#Save Model
model.save('fruitsmodel.h5')
```

Model Tuning:

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
early_stop = EarlyStopping(monitor='val_accuracy', 
                           patience=5)

lr = ReduceLROnPlateau(monitor='val_accuracy', 
                       factor=0.5, 
                       min_lr=0.00001)

callback = [early_stop,lr]
```

```python
# Train model

model.fit_generator(ftrain,
                    steps_per_epoch=len(ftrain),
                    epochs=100,
                    callbacks=callback,
                    validation_data=ftest,
                    validation_steps=len(ftest))
```
